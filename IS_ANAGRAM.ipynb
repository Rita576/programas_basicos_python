```
def is_anagram(s1,s2):
    lista = list(s2)
    pos1 = 0
    certo = True


    while pos1 < len(s1) and certo:
        pos2 = 0
        posicao = False
        while pos2 < len(lista) and not posicao:
            if s1[pos1] == lista[pos2]:
                posicao = True
            else:
                pos2 = pos2 + 1
        if posicao:
            lista[pos2] = None
        else:
            certo = False
            pos1 = pos1 + 1
        return certo
print(is_anagram('abcd','dcba'))
print(is_anagram('amor','roma'))



```
